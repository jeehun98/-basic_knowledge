```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-5, 5, 0.01)
```

```python
def gaussian(x, mean, sigma):
  return (1 / np.sqrt(2 * np.pi * sigma**2)) * np.exp(-(x-mean)**2 / (2 * sigma**2))
```

```python
legend = []

plt.plot(x, gaussian(x, 0, 1))

plt.show
```

```python
plt.plot(x, gaussian(x, 0, 2))

plt.show
```
